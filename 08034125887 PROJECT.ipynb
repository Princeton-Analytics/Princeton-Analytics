```python
# Loop through columns to round the float-point numbers to 2 decimals to make the data cleaner and more readable.
for i in range(len(df.columns)):
    if i < 2:
        i + 1
    else:
        column = df.columns[i]
        df[column] = df[column].round(2)

# Ensure the Year column doesn't have any decimals
df['Year'].round(0)

df.head(3)
```

```python
# Loop through columns to clean up null values
for i in range(len(df.columns)):
    if i < 3:
        i + 1
    else:
        column = df.columns[i]
        df[column].fillna(0,inplace=True)
        
# Confirm null values are cleaned up
df.isnull().sum()
```
